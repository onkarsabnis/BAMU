```python
from __future__ import with_statement
import matplotlib
matplotlib.use('TkAgg')

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure


import numpy as np
import tkinter as tk
from tkinter import *
from tkinter import font
from tkinter import filedialog
from tkinter.filedialog import askopenfilename
from pylab import *
from tkinter import Button
import asciitable
import csv
import pandas as pd

import qgrid
#qgrid.nbisntall()

from pandastable import Table, TableModel
```

```python
class SLI:
    def __init__(self, master):
        self.master = master        
        master.title("Spectral Library Interface for NISA HyperSpectral Analysis Tool")
        menu = Menu(master)
        master.config(menu=menu)

        filemenu = Menu(menu)
        menu.add_cascade(label="File", menu=filemenu)

        filemenu.add_cascade(label="ASCII Files", command=self.endmember)
        filemenu.add_cascade(label="ASD Filedspec4 File", command=self.endmember)
        filemenu.add_command(label="ASTER Database", command=self.endmember)
        filemenu.add_command(label="NISA Spectral Profiles", command=self.endmember)
        filemenu.add_separator()

        databasemenu = Menu(menu)
        menu.add_cascade(label="Database", menu=databasemenu)

        databasemenu.add_cascade(label="ASTER Database", command=self.endmember)
        databasemenu.add_cascade(label="USGS Database", command=self.endmember)
        databasemenu.add_cascade(label="NISA Database", command=self.endmember)
        
     
        self.lb1 = Label(master,text="File Information")
        self.lb1.grid(row=0, column =0)

        self.lb2 = Label(master,text="Spectral Plot")
        self.lb2.grid(row=0, column =6)

        self.lb3 = Label(master,text="Spectral Information")
        self.lb3.grid(row=0, column =12)

        self.list1 =Listbox(master, height=28, width=68)
        self.list1.grid(row=2,column=0,rowspan =6, columnspan=2)

        self.list2 =Listbox(master, height=28, width=68)
        self.list2.grid(row=2,column=6,rowspan =6, columnspan=2)

        self.list3 =Listbox(master, height=28, width=60)
        self.list3.grid(row=2,column=12,rowspan =6, columnspan=2)

        self.sb1 = Scrollbar(master)
        self.sb1.grid(row=2, column=18, rowspan=6)
        self.sb1.config(command=self.list3.yview)
        self.list3.config(yscrollcommand = self.sb1.set)

        self.b3 = Button(master, borderwidth=5, width=10, text="Clear Screen", command=self.ClearScreen)
        #self.b3.pack()
        self.b3.place(x=182, y=420)

        self.b4 = Button(master, borderwidth=5, width=10, text="Save", command=self.ClearScreen)
        #self.b4.pack()
        self.b4.place(x=580, y=420)

    def ClearScreen(self):
        self.list1.delete(0, END)
        self.list2.delete(0, END)
        self.list3.delete(0, END)
        self.canvas.get_tk_widget().delete("all")
        
    def endmember(self):
        filename = askopenfilename()
        if filename.endswith('txt'):
            usecols = (0, 1)
            header = {}

            with open(filename, 'rt') as f:
                for header_lines, line in enumerate(f):
                    line = line.strip()
                    if not line: break # assume that a blank line means "header stops here"
                    
                    key, value = line.split(' ', 1)
                    header[key] = value
                    self.list1.insert(1, value)
                    

            depth, N = np.loadtxt(filename, skiprows=header_lines + 2, unpack=True, usecols=usecols)
            
            df = pd.DataFrame({'Wavelength':depth,'Reflectance':N})
            print(df)
            Row_list =[]
            my_list=[]
            # Iterate over each row 
            for columns in df.itertuples(index=False):
                # Create list for the current row
                #my_list_ =[columns.Wavelength, columns.Reflectance]
                #m=np.array(columns.Wavelength)
                #print(m)
                #my_list_df =pd.DataFrame(data=[columns.Wavelength.reshape(-1, 2), columns.Reflectance.reshape(-1, 2)],columns=['Wavelength','Reflectance'])
                print(columns)
                # append the list to the final list
                #Row_list.append('\n'.join(map(str, my_list)))
##                Row_list.append(depth)
                                         
            #print(Row_list)
            #my_list=np.array(my_list)
            #print(my_list_df)                            
            #Row_list=np.array(Row_list)
            #datafrrame=pd.DataFrame(Row_list,columns=['ROW_LIST'])
            #print(datafrrame)
            a=np.linspace(490,0,491)
            a=a.astype(int)
            #self.list3.insert(1,'_______________________________')
            #self.list3.insert(1,'\n')
            #self.list3=np.array(self.list3)
            self.list3.insert(1,'Wavelength      Reflectance')
            
            #self.list3.insert(3,'Reflectance')
            for i in a:
                
                self.list3.insert(1,df['Reflectance'][i])
                self.list3.insert(1,df['Wavelength'][i])
                #self.list3.insert(4,'\t')
                
                #self.list3.insert(4,'\n')
                self.list3.insert(1,'\n')
            
            
            
            fig = Figure(figsize=(4,4))
            a = fig.add_subplot(111)
            a.plot(depth, N,color='blue')
            a.set_title ("Spectral Graph", fontsize=14)
            a.set_ylabel("Reflectance(%)", fontsize=12)
            a.set_xlabel("Wavelength(Micrometer)", fontsize=12)

            self.canvas = FigureCanvasTkAgg(fig, master=self.master)
            self.canvas.get_tk_widget().place(x = 425,y = 24)
            self.canvas.draw()          
        else:
            self.list1.insert(0, 'File type not supported')    
```
