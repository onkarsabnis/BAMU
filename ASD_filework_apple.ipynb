```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
DF=pd.read_csv('ASD_apple.csv')
```

```python
DF.info()
```

```python
DF.columns.tolist()
```

```python
a=np.linspace(0,236611,111)

a=a.astype(int)
```

```python
#(np.linspace(0,236611,111)).astype(int)
```

```python
df=list()

for i in a:
    df.append(DF.iloc[i:i+2151])
```

```python
def NDVI(i):
    return ((df[i].loc[df[i]['Wavelength'] == 800, 'Reflectance'].iloc[0])-(df[i].loc[df[i]['Wavelength'] == 670, 'Reflectance'].iloc[0]))/((df[i].loc[df[i]['Wavelength'] == 800, 'Reflectance'].iloc[0])+(df[i].loc[df[i]['Wavelength'] == 670, 'Reflectance'].iloc[0]))
    
```

```python
ndvi=[]
for i in range(0,110):
    m= NDVI(i)
    ndvi.append(m)
```

```python
def NDNI(i):
    return ((-np.log(df[i].loc[df[i]['Wavelength'] == 1510, 'Reflectance'].iloc[0]))+(np.log(df[i].loc[df[i]['Wavelength'] == 1680, 'Reflectance'].iloc[0])))/((-np.log(df[i].loc[df[i]['Wavelength'] == 1510, 'Reflectance'].iloc[0]))-(np.log(df[i].loc[df[i]['Wavelength'] == 1680, 'Reflectance'].iloc[0])))
    
    
```

```python
ndni=[]
for i in range(0,110):
    m= NDNI(i)
    ndni.append(m)
```

```python
def ARI1(i):
    return (1/(df[i].loc[df[i]['Wavelength'] == 550, 'Reflectance'].iloc[0]))-(1/(df[i].loc[df[i]['Wavelength'] == 700, 'Reflectance'].iloc[0]))
    
```

```python
ari1=[]
for i in range(0,110):
    m= ARI1(i)
    ari1.append(m)
```

```python
def ARI2(i):
    return ((df[i].loc[df[i]['Wavelength'] == 800, 'Reflectance'].iloc[0])*(1/(df[i].loc[df[i]['Wavelength'] == 550, 'Reflectance'].iloc[0]))-(1/(df[i].loc[df[i]['Wavelength'] == 700, 'Reflectance'].iloc[0])))
    
    
```

```python
ari2=[]
for i in range(0,110):
    m= ARI2(i)
    ari2.append(m)
```

```python
def CRI1(i):
    return (1/(df[i].loc[df[i]['Wavelength'] == 510, 'Reflectance'].iloc[0]))-(1/(df[i].loc[df[i]['Wavelength'] == 550, 'Reflectance'].iloc[0]))
    
    
```

```python
cri1=[]
for i in range(0,110):
    m= CRI1(i)
    cri1.append(m)
```

```python
def CRI2(i):
    return (1/(df[i].loc[df[i]['Wavelength'] == 510, 'Reflectance'].iloc[0]))-(1/(df[i].loc[df[i]['Wavelength'] == 700, 'Reflectance'].iloc[0]))
    
    
```

```python
cri2=[]
for i in range(0,110):
    m= CRI2(i)
    cri2.append(m)
```

```python
ndni=np.array(ndni)
ari1=np.array(ari1)
ari2=np.array(ari2)
cri1=np.array(cri1)
cri2=np.array(cri2)
ndvi=np.array(ndvi)
```

```python
columns=['ndni','ari1','ari2','cri1','cri2','ndvi']

dfnnew=pd.DataFrame([ndni,ari1,ari2,cri1,cri2,ndvi],columns)
```

```python
dfnnew
```

```python
dfnnew_trans=dfnnew.T
```

```python
sns.lmplot('ndvi','ndni',data=dfnnew_trans)
```

```python
from scipy import stats
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(dfnnew_trans['ndvi'],dfnnew_trans['ndni'])

# use line_kws to set line label for legend
ax = sns.regplot(x="ndvi", y="ndni", data=dfnnew_trans, color='b', 
 line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)})

# plot legend
ax.legend()

plt.show()

#enter image description here
```
